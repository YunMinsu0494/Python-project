```python
# crossTab: 내가 원하는 행과 열을 설정해서 빈도값을 보는것
import pandas as pd
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
pd.crosstab(index = tips["sex"], columns = tips["day"], values = tips["tip"], aggfunc = "sum",
            margins = True, margins_name = "총계", normalize = "columns")
```

```python
# Pivot Table: 내가 원하는 행, 열을 설정해서 직계값을 보는것
import warnings
warnings.filterwarnings("ignore")

tips.pivot_table(index = "sex", columns = "day", values = ["tip", "total_bill"], aggfunc = "mean")
```

```python
tips.pivot_table(index = ["sex", "time"], columns = "day", values = "tip", aggfunc = ["sum", "mean"],
                fill_value = 0, margins = True, margins_name = "총계")
```

```python
tips.pivot_table(index = ["sex", "time"], columns = "day", values = ["tip", "total_bill"], aggfunc = {"tip" : "mean", "total_bill" : "sum"},
                fill_value = 0, margins = True, margins_name = "총계")
```
