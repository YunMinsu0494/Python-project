```python
import pandas as pd
pew = pd.read_csv("Data/pew.csv")

# 열에 있는 값들을 행으로 내릴 수 있음
pd.melt(pew, id_vars = "religion", var_name = "income", value_name = "count")
```

```python
# 퀴즈: billboard.csv // 음악 점수 데이터 적절하게 melting 하기
billboard = pd.read_csv("Data/billboard.csv")
pd.melt(billboard, id_vars = ["year","artist","track","time","date.entered"], var_name = "week", value_name = "rating_score")
```

```python
# country_timeseries.csv // 에볼라 바이러스 생존자, 사망자 수
ebola = pd.read_csv("Data/country_timeseries.csv")
ebola2 = pd.melt(ebola, id_vars = ["Date","Day"])

case = ebola2["variable"].str.split("_").str[0] # 이 시리즈에 접근한 후 _ 기준으로 리스트화 한다
country = ebola2["variable"].str.split("_").str[1]

ebola2["Case"] = case
ebola2["Country"] = country
del ebola2["variable"]
ebola2
```

```python
# weather.csv // 날씨 최저온도.최고온도
weather = pd.read_csv("Data/weather.csv")
weather2 = pd.melt(weather, id_vars = ["id","year","month","element"], var_name = "Day", value_name = "Temp")

# 내가 원하는 행과, 열을 설정해서 집계값
weather2.pivot_table(index = ["year","month","Day"],
                    columns = "element",
                    values = "Temp", dropna = False)
```

```python
# .str 접근법
s = pd.Series(["  Apple  ","  Banana  ","  Cherry  "])
# s.str.upper()
# s.str.len() 길이 확인
# s.str.contains("an") 입력한 값이 포함이 되어있는지
# s.str.startswith(" ") 입력한 값으로 시작하는지 
# s.str.endswith("na") 입력한 값으로 끝나는지
# s.str.split(" ") 입력한 값을 기준으로 자르기

s = pd.Series([["2025","12","13"],["2025","03","01"]])
s.str.join("-")
```
