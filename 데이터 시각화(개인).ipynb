```python
# 데이터 시각화가 왜 필요한지?
import pandas as pd
import seaborn as sns # 자동으로 예쁜(논문에 올려도 될 정도) 그래프 // 연습용 데이터
import matplotlib.pyplot as plt #데이터 시각화 기반
```

```python
anscombe = sns.load_dataset("anscombe")

data1 = anscombe[anscombe["dataset"] == "I"] # anscombe에서 dataset이 I인 것들만 가져기기
data2 = anscombe[anscombe["dataset"] == "II"]
data3 = anscombe[anscombe["dataset"] == "III"]
data4 = anscombe[anscombe["dataset"] == "IV"]

print(data1["x"].mean(), data1["y"].mean())
print(data2["x"].mean(), data2["y"].mean())
print(data3["x"].mean(), data3["y"].mean())
print(data4["x"].mean(), data4["y"].mean())
```

```python
print(data1["x"].std(), data1["y"].std())
print(data2["x"].std(), data2["y"].std())
print(data3["x"].std(), data3["y"].std())
print(data4["x"].std(), data4["y"].std())
```

```python
print(data1["x"].corr(data1["y"])) #corr: 상관관계
print(data2["x"].corr(data2["y"]))
print(data3["x"].corr(data3["y"]))
print(data4["x"].corr(data4["y"]))
```

```python
fig = plt.figure() #흰색 도화지

ax1 = fig.add_subplot(2, 2, 1) #2행 2열을 1번째 리리
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.scatter(data1["x"], data1["y"]) #산점도(점 그래프)
ax2.scatter(data2["x"], data2["y"])
ax3.scatter(data3["x"], data3["y"])
ax4.scatter(data4["x"], data4["y"])

ax1.set_title("Data1") #제목 설정
ax2.set_title("Data2")
ax3.set_title("Data3")
ax4.set_title("Data4")

fig.suptitle("Anscombe Data") #메인 제목 설정
fig.tight_layout() #그림에서 안맞는 부분을 알아서 해결해서 그려라

plt.savefig("Anscombe.png") #파일을 png파일로 저장
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# 일변량 그래프(연속형 데이터): 히스토그램 // 이상치 탐지
sns.set_style("whitegrid") #white, whitegrid, dark, darkgrid
sns.histplot(data = tips, x = "total_bill", bins = 20, kde = True) #bins = 막대기를 20개로 늘려라 // kde = True: 밀도함수
plt.savefig("hist.png")
plt.show()
```

```python
# 이변량 그래프(둘다 연속형): 산점도
sns.lmplot(data = tips, x = "total_bill", y = "tip", fit_reg = False) # fit_reg = False: 추세선 삭제
sns.kdeplot(data = tips, x = "total_bill", y = "tip")
plt.show()
```

```python
# 확률 밀도함수 그래프(이변량)
sns.kdeplot(data = tips, x = "total_bill", y = "tip")
plt.show()
```

```python
# 다변량 그래프
sns.lmplot(data = tips, x = "total_bill", y = "tip", hue = "sex",  fit_reg = False, col = "time", row = "smoker")
plt.show()
```

```python
# 산점도 + 히스토그램
sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "hex") #kind = "hex": 산점도를 육각형으로
plt.show()
```

```python
# 일변량 그래프(범주형 데이터): 빈도 막대 그래프
sns.countplot(data = tips, x = "day")
plt.show()
```

```python
# 평균 막대 그래프(이변량 그래프) 범주형, 연속형
sns.barplot(data = tips, x = "sex", y = "tip")
plt.show()
```

```python
# box plot: 이상치 탐지
sns.boxplot(data = tips, x = "sex", y = "tip")
plt.show()
```

```python
sns.violinplot(data = tips, x = "sex", y = "tip")
sns.swarmplot(data = tips, x = "sex", y = "tip", color = "white")
plt.show()
```

```python
# 산점도 행렬
sns.pairplot(tips)
plt.show()
```
