```python
import pandas as pd
import numpy as np

df1=pd.read_csv("Data/concat_1.csv")
df2=pd.read_csv("Data/concat_2.csv")
df3=pd.read_csv("Data/concat_3.csv")
```

```python
df2
```

```python
df=pd.concat([df1,df2,df3], ignore_index=True, axis=0) #기본값 axis=0 // ignore_index=True: 합칠떄 열이름을 상관하지 않고 순서대로 배열
df
```

```python
df2.columns=["A","C","E","F"]
df2
pd.concat([df1,df2,df3], join="inner") #공통된 열만 합칠때, *기본값은 outter
```

```python
df=pd.concat([df1,df2,df3],axis=1, ignore_index=True)
df
```

```python
df2.index=[0,2,4,6]
df2
pd.concat([df1,df2,df3],axis=1 ,join="inner") #공통된 행만 합칠때, *기본값은 outer
```

```python
pd.concat([df1,df2,df3], keys=["df1","df2","df3"])
```
