```python
df=pd.read_excel("Data/성적 처리.xlsx",sheet_name="Sheet1")
#axis가 0이면 아래방향, 1이면 오른쪽 방향
df["총점"]=df.iloc[:,2:].sum(axis=1)
df["평균"]=df["총점"]/5
df
```

```python
df.sort_values("평균") #기본은 오름차순 정렬
```

```python
df.sort_values("평균",ascending=False) #내림차순 정렬
```

```python
#특성 정렬
df.sort_values(["반","평균"],ascending=[True,False])
```

```python
#행 정렬
df.sort_index() #기본적으로 오름차순 정렬
```

```python
df.sort_index(ascending=False)
```

```python
#문자열/사전순으로 정렬
df.sort_index(axis=1)
```

```python
#순위 구하기 // 3등이 2명이면 둘다 3.5등으로 처리하고 다음 등수를 5등으로 처리 => 4등이 x
df["순위"]=df["평균"].rank(ascending=False, method="average")
df
```

```python
df["순위"]=df["평균"].rank(ascending=False, method="max")
df
```

```python
#3등이 2명이면 둘다 3등으로 하고 다음 등수를 5등으로 처리.
df["순위"]=df["평균"].rank(ascending=False, method="min")
df
```

```python
#3등이 2명이면, 먼저 발견한 3등은 3등, 다음 발견된 3등은 4등 => 선착순 프로그램에서 활용
df["순위"]=df["평균"].rank(ascending=False, method="first")
df
```

```python
#min과 동일
df["순위"]=df["평균"].rank(ascending=False, method="dense")
df
```

```python
df["순위"]=df["평균"].rank(ascending=False, method="dense", pct=True) #백분률로 변환
df
```
